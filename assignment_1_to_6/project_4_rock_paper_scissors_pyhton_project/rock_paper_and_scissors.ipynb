#Project 4: Rock, paper, scissors Python Project


Introduction:
"Rock, Paper, Scissors" is a fun and simple game where the player competes against the computer. The game follows basic rules:

Rock beats Scissors (Rock crushes Scissors)

Scissors beats Paper (Scissors cuts Paper)

Paper beats Rock (Paper covers Rock)

The player selects one option (Rock, Paper, or Scissors), and the computer randomly chooses one too. The winner is decided based on the rules above.


How It Works:

The program starts and asks the user to choose Rock, Paper, or Scissors.
 The computer randomly selects Rock, Paper, or Scissors.
The program compares both choices:

If both are the same → It's a Tie!

If the user's choice beats the computer's → User Wins!

If the computer's choice beats the user's → Computer Wins!

The game repeats until the user types "exit" to quit.



```python
import random

def main():
    print("Welcome to Rock, Paper, Scissors!")

    choices = ["rock", "paper", "scissors"]

    while True:
        user_choice = input("Choose Rock, Paper, or Scissors (or type 'exit' to quit): ").lower()

        if user_choice == "exit":
            print("Thanks for playing! Goodbye.")
            break

        if user_choice not in choices:
            print("Invalid choice. Please choose Rock, Paper, or Scissors.")
            continue

        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        if user_choice == computer_choice:
            print("It's a tie!")

        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("You win!")
        else:
            print("Computer wins!")

        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            print("Thanks for playing! Goodbye.")
            break

if __name__ == "__main__":
    main()
```
