# Guess the Number Game (User) Python Project

Introduction:

This is a fun number guessing game where the computer selects a random number between 1 and 100, and the user tries to guess the correct number. The game provides hints like "Too High!" or "Too Low!" to help the player.

How It Works:

The computer randomly selects a number between 1 and 100.
The user has 5 attempts to guess the correct number.
 After each guess, the computer gives feedback:

"Too high!" → If the guess is greater than the number.

"Too low!" → If the guess is smaller than the number.
 If the user guesses correctly, they win the game.
 If they run out of guesses, the game ends and the computer reveals the correct number.



```python
import random

def main():
    print("Welcome to the Guess the Number Game!")
    print("I have chosen a number between 1 and 100.")
    print("Try to guess it!")

    random_number = random.randint(1, 100)
    # print(random_number)
    guess = 5

    while guess > 0:
        user_guess = int(input("Enter your guess: "))

        if user_guess == random_number:
            print("Congratulations! You guessed the number!")
            break
        elif user_guess > random_number:
            print("Your guess is too high!")
        else:
            print("Your guess is too low!")

        guess -= 1

        if guess == 0:
            print(f"Sorry, you ran out of guesses. The number was {random_number}.")

if __name__ == "__main__":
    main()
```
