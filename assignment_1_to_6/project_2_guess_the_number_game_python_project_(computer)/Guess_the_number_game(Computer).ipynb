# Guess the Number Game Python Project (Computer)

Introduction:

This is a fun Python game where the user thinks of a secret number, and the computer tries to guess it. After each guess, the user provides feedback saying whether the guess is "Too High", "Too Low", or "Correct".

How It Works:

The user thinks of a number between 1 and 10.

The computer makes a guess.

The user provides feedback:

If the guess is too high → "too high"

If the guess is too low → "too low"

If the guess is correct → "correct"

The computer intelligently adjusts its next guess.

The game continues until the computer guesses correctly.



```python
import random

def computer_guesses():
    print("\n Welcome to the Number Guessing Game (Computer)!")
    print("Think of a number between 1 and 10, and I'll try to guess it.")
    print("After each guess, tell me if it's 'high', 'low', or 'correct'.\n")

    low = 1
    high = 10
    attempts = 0

    while True:
        guess = random.randint(low, high)
        attempts += 1
        print(f"My guess is: {guess}")

        feedback = input("Is this too high, too low, or correct? ").lower()

        if feedback == 'correct':
            print(f"Yay! I guessed your number in {attempts} attempts!")
            break
        elif feedback == 'high':
            high = guess - 1
        elif feedback == 'low':
            low = guess + 1
        else:
            print("Please enter 'high', 'low', or 'correct'.")

        if low > high:
            print("Hmm, something doesn't add up. Did you change your number? ")
            break

if __name__ == "__main__":
    computer_guesses()
```
