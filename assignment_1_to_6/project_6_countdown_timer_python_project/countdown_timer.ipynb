# Project 5: Countdown Timer Python Project

Introduction: A Countdown Timer is a program that counts down from a given time (in seconds) to zero, commonly used in alarms, exams, and events.

 Working:
User enters time (in seconds).

Timer converts seconds into minutes:seconds format using divmod().

Updates every second using time.sleep(1).

Displays real-time countdown using sys.stdout.write().

Stops at 00:00 and prints "Time's up!".




```python
import time
import sys

def countdown(seconds):
    print("Countdown Timer")
    print("----------------")

    while seconds >= 0:
        minutes, remaining_seconds = divmod(seconds, 60)
        timer_display = f"\r{minutes:02d}:{remaining_seconds:02d}"

        sys.stdout.write(timer_display)
        sys.stdout.flush()

        time.sleep(1)
        seconds -= 1

    print("\nTime's up!")

while True:
    try:
        time_in_seconds = int(input("Enter the time in seconds: "))
        if time_in_seconds < 0:
            print("Please enter a non-negative number.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a valid number.")

countdown(time_in_seconds)
```
