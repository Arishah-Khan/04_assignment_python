# Project 6: Password Generator Python Project

 Introduction:
A Password Generator is a program that creates strong, random passwords for security purposes. It helps users generate unique passwords of desired length and complexity, preventing weak or guessable passwords.

 Working:
Takes user input → Asks how many passwords are needed and their length.

Defines character set → Uses uppercase, lowercase, numbers, and symbols.

Randomly selects characters → Uses random.choice(chars) to pick characters.

Creates passwords → Joins selected characters into a string.

Prints the generated passwords → Displays them for the user.




```python
import random

def password_generator():

    print("Welcome to the password generator")
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*().,?0123456789"

    num_passwords = int(input("How many passwords do you need? "))
    password_length = int(input("Enter the length of each password: "))

    print("Here are your password")

    for pwd in range(num_passwords):
        passwords = ""
        for char in range(password_length):
            passwords += random.choice(chars)

        print(passwords)

password_generator()
```
