#Project 5: Hangman Python Project

Start the game – The computer selects a random word.

Guess a letter – Type a single letter and press Enter.

Correct guess – The letter appears in the word.

Wrong guess – The letter does not appear, and you lose a life.

Continue guessing – Keep guessing until you complete the word or lose all lives.

Win the game – If you guess all letters correctly, you win!

Lose the game – If you run out of lives before guessing the word, you lose.

```python
import random

def main():
    print("Welcome to Hangman!")
    print("-------------------")

    words = ["apple", "mango", "orange", "banana", "grapes", "peach"]
    word_to_guess = random.choice(words)
    hidden_word = ["_"] * len(word_to_guess)
    lives = 6

    print(" ".join(hidden_word))

    while lives > 0 and "_" in hidden_word:
        user_guess = input("Guess a letter: ").lower()

        if user_guess in word_to_guess:
            print("Correct!")
            for i in range(len(word_to_guess)):
                if word_to_guess[i] == user_guess:
                    hidden_word[i] = user_guess
        else:
            lives -= 1
            print(f"Wrong guess! {lives} lives left.")

        print(" ".join(hidden_word))

    if "_" not in hidden_word:
        print("Congratulations! You win")
    else:
        print(f"You lost! The word was: {word_to_guess}")

if __name__ == "__main__":
    main()


```
