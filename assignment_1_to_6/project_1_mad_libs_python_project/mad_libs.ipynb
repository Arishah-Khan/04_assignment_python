# Mad Libs Python Project

Introduction:

Mad Libs is a fun word game where users provide different types of words (adjectives, nouns, verbs, etc.), and these words are inserted into a pre-written story. Every time the user enters different words, a unique and funny story is generated!

```
# This is formatted as code
```



How It Works:
Take input from the user (adjective, noun, verb, etc.)

Define a story format

Insert user inputs into the story

Display the final story


````python
def main():

    adjective = input("Enter an adjective (e.g., funny, brave, sleepy,silly): ")
    verb = input("Enter a verb (action word, e.g., run, jump, dance): ")
    noun = input("Enter a noun (a thing, e.g., book, tree, hat , banana , table): ")
    animal = input("Name an animal: ")
    place = input("Enter a place (e.g., park, school, castle , jungle , forest): ")
    emotion = input("Enter an emotion (e.g., happy, scared, excited): ")
    past_sentence_verb = input("Enter a past-tense verb (e.g., laughed, cried, ran): ")

    mad_libs = f"One day, a very {adjective} {animal} decided to {verb} in the {place}. \
It was a {adjective} idea because suddenly, a {noun} appeared!  \
The {animal} got so {emotion} that it started to {verb} quickly. \
In the end, everyone {past_sentence_verb} and lived happily ever after!"

    print("\n Your Mad Libs Story ")
    print(mad_libs)

if __name__ == "__main__":
    main()
````
