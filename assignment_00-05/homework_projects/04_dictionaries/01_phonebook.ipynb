# Problem Statement  

In this program we show an example of using dictionaries to keep track of information in a phonebook.

```python
phonebook = {}

def add_entry(name, number):
    phonebook[name] = number
    print(f"Added: {name} -> {number}")

def lookup_entry(name):
    number = phonebook.get(name)
    if number:
        print(f"{name}'s number is {number}")
    else:
        print(f"{name} not found in phonebook.")

def delete_entry(name):
    if name in phonebook:
        del phonebook[name]
        print(f"{name} deleted from phonebook.")
    else:
        print(f"{name} not found in phonebook.")

def read_entries():
    if phonebook:
        print("\nPhonebook Entries:")
        for name, number in phonebook.items():
            print(f"{name}: {number}")
    else:
        print("Phonebook is empty.")

def main():
    while True:
        print("\nPhonebook Menu:")
        print("1. Add Entry")
        print("2. Lookup Entry")
        print("3. Delete Entry")
        print("4. Read All Entries")
        print("5. Exit")

        try:
            user = int(input("Choose an option (1-5): "))
        except ValueError:
            print("Invalid input! Please enter a number between 1 and 5.")
            continue

        if user == 1:
            name = input("Enter name: ")
            number = input("Enter number: ")
            add_entry(name, number)

        elif user == 2:
            name = input("Enter name: ")
            lookup_entry(name)

        elif user == 3:
            name = input("Enter name to delete: ")
            delete_entry(name)

        elif user == 4:
            read_entries()

        elif user == 5:
            print("Exiting phonebook. Goodbye!")
            break

        else:
            print("Invalid choice! Please choose between 1 and 5.")

if __name__ == "__main__":
    main()
```
