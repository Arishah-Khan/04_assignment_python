# Problem Statement  


Simulate rolling two dice, and prints results of each roll as well as the total.

```python
import random

def main():
    dice_number = 6

    roll_1 = random.randint(1, dice_number)
    roll_2 = random.randint(1, dice_number)

    total = roll_1 + roll_2

    print(f"Roll 1: {roll_1}")
    print(f"Roll 2: {roll_2}")
    print(f"Total: {total}")

if __name__ == "__main__":
    main()


```
