# Problem Statement  

Write a function that takes two numbers and finds the average between the two.





```python
def find_average(num1, num2):
    result = (num1 + num2) / 2
    return result

def main():
    result1 = find_average(10, 4)
    print(result1)

    result2 = find_average(20, 5)
    print(result2)

if __name__ == "__main__":
    main()


```
